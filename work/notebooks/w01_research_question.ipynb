False Negative:
A page with strong AI referral potential may be ignored, causing the business to miss opportunities for additional traffic and visibility.

## 3. Quick look at the data (2-3 real numbers)

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/hUSsAin976-tech/ML-internship-at-FlyRank/main/data/raw/content_refresh_anonymized.csv")

print("Total Pages:", len(df))

print("Average CTR:", round(df["ctr"].mean(), 4))

print("Average Position:", round(df["avg_position"].mean(), 2))

print("Average Word Count:", round(df["word_count"].mean(), 0))

if "ai_sessions_90d" in df.columns:
    print("Pages with AI Sessions:",
          (df["ai_sessions_90d"] > 0).sum())

if "ai_traffic_pct" in df.columns:
    print("Average AI Traffic Percentage:",
          round(df["ai_traffic_pct"].mean(),4))
```
